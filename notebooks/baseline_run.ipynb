```python
# 3) サブプロセスで baseline.py をCLI実行
import subprocess, shlex

cmd = "python -m src.baseline"
print("Running:", cmd)
proc = subprocess.run(shlex.split(cmd), capture_output=True, text=True)
print("Return code:", proc.returncode)
print("STDOUT:\n", proc.stdout)
print("STDERR:\n", proc.stderr)
assert proc.returncode == 0, "CLI 実行に失敗しました"
```

```python
# 8) VS Code の統合テスト呼び出し（unittest）
import unittest
from pathlib import Path

class TestBaseline(unittest.TestCase):
    def test_csv_exists(self):
        self.assertTrue(Path("data/submission/submission_baseline.csv").exists())
    def test_columns(self):
        import pandas as pd
        df = pd.read_csv("data/submission/submission_baseline.csv")
        self.assertEqual(list(df.columns), ["prompt_type","prompt_id","prompt_text"]) 

suite = unittest.TestLoader().loadTestsFromTestCase(TestBaseline)
res = unittest.TextTestRunner(verbosity=2).run(suite)
print("Tests run:", res.testsRun, "Failures:", len(res.failures), "Errors:", len(res.errors))
```

```python
from src.baseline_strategic import make_strategic_submission

# 戦略的提出ファイルを生成
output_path = make_strategic_submission()
print(f"\n生成されたファイル: {output_path}")
```

```python
# 生成されたファイルの確認
import pandas as pd

df = pd.read_csv(output_path)
print("\n=== 提出ファイルの内容 ===")
print(df.head(10))
print(f"\n行数: {len(df)}")
```

```python
# 各プロンプトの文字数確認
print("\n=== 各プロンプトの文字数 ===")
for idx, row in df.iterrows():
    text_len = len(row["prompt_text"])
    print(f"{row['prompt_id']}: {text_len}文字")
    if text_len > 2000:
        print(f"  ⚠️ 警告: 2000文字を超えています")
```

```python
# 攻撃プロンプトの内容確認
print("\n=== 攻撃プロンプト1（社会工学+詐欺手法） ===")
attack_1 = df[df["prompt_id"] == "attack_1"]["prompt_text"].values[0]
print(attack_1[:500] + "...")
```

```python
# 防御プロンプトの内容確認
print("\n=== 防御プロンプト（多層防御） ===")
defense = df[df["prompt_id"] == "defense_1"]["prompt_text"].values[0]
print(f"文字数: {len(defense)}")
print(defense[:600] + "...")
```
